```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"D:\DS\Housing.csv")
```

```python
plt.hist(df['price'], bins=30)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df['area'], df['price'])
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
avg_bed = df.groupby('bedrooms')['price'].mean()

avg_bed.plot(kind='bar')
plt.title("Average Price by Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price")
plt.show()
```

```python
df.groupby('bathrooms')['price'].mean().plot(kind='bar')
plt.title("Average Price by Bathrooms")
plt.xlabel("Bathrooms")
plt.ylabel("Average Price")
plt.show()
```

```python
df.groupby('furnishingstatus')['price'].mean().plot(kind='bar')
plt.title("Price by Furnishing Status")
plt.xlabel("Furn. status")
plt.ylabel("Average Price")
plt.show()
```

```python
df.groupby('parking')['price'].mean().plot(kind='line', marker='o')
plt.title("Parking vs Average Price")
plt.xlabel("Parking Spaces")
plt.ylabel("Average Price")
plt.show()
```

```python
df.groupby('airconditioning')['price'].mean().plot(kind='bar')
plt.title("Air Conditioning Impact on Price")
plt.show()
```

```python
df.groupby('prefarea')['price'].mean().plot(kind='bar')
plt.title("Preferred Area vs Price")
plt.show()
```

```python
numeric_df = df[['price','area','bedrooms','bathrooms','stories','parking']]
corr = numeric_df.corr()

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=45)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```
